```python
table_start = int(input('Enter your table start '))
table_end = int(input('Enter your table till '))
table_length = int(input('Enter your table length'))


for method in range (table_start, table_end):
    print('table of {method}')
    for temp_2 in range (table_length+2):
        print(f"{method} * {temp_2} = {method * temp_2}")
```
