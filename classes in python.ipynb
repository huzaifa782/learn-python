```python
class student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def get_avg(self):
        sum = 0
        for val in self.marks:
            sum += val
        print("hi", self.name, "your avg marks is: ", sum/3)
```

```python
s1 = student("huzaifa", [99, 98, 88])
```

```python
s1.marks
```

```python
s1.get_avg
```

```python
s1.name
```
